```python
import numpy as np 
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
import math
```

```python
df = pd.read_excel('results.xlsx')
list(df.columns)

df['Spawn Rate'].unique()
```

```python
p1 = df.loc[ df['Player 1'] == 8 ]
p2 = df.loc[ df['Player 2'] == 8 ]
p3 = df.loc[ df['Player 3'] == 8 ]
p4 = df.loc[ df['Player 4'] == 8 ]
g8 = pd.concat([p1, p2, p3, p4])
```

```python
p1 = p1.sort_values(by=['Total Day'], ascending=True)
p2 = p2.sort_values(by=['Total Day'], ascending=True)
p3 = p3.sort_values(by=['Total Day'], ascending=True)
p4 = p4.sort_values(by=['Total Day'], ascending=True)

sc1 = p1.sort_values(by=['Total Day', 'Player 1 Final Score'], ascending=True)
sc2 = p2.sort_values(by=['Total Day', 'Player 2 Final Score'], ascending=True)
sc3 = p3.sort_values(by=['Total Day', 'Player 3 Final Score'], ascending=True)
sc4 = p4.sort_values(by=['Total Day', 'Player 4 Final Score'], ascending=True)
```

```python
fig = plt.subplots()

players = [1, 2, 3, 4] # x
suc = [22, 10, 12, 24]
fal = [12, 8, 13, 24]

x_ax = np.arange(len(players))

plt.bar(x_ax - 0.2, suc, 0.4, label = "Success")
plt.bar(x_ax + 0.2, fal, 0.4, label = "Failure")

plt.xticks(x_ax, ["Player 1", "Player 2", "Player 3", "Player 4"])
plt.title("Number of Timeouts vs No Timeouts For Group 8")
plt.xlabel('Players')
plt.ylabel('Runs')
plt.legend()
plt.show()
```

```python
p1_50   =   sc1.loc[ sc1['Total Day'] == 50   ]
p1_100  =   sc1.loc[ sc1['Total Day'] == 100  ]
p1_500  =   sc1.loc[ sc1['Total Day'] == 500  ]
p1_1000 =   sc1.loc[ sc1['Total Day'] == 1000 ]
p1_2000 =   sc1.loc[ sc1['Total Day'] == 2000 ]

p2_50   =   sc2.loc[ sc2['Total Day'] == 50   ]
p2_100  =   sc2.loc[ sc2['Total Day'] == 100  ]
p2_500  =   sc2.loc[ sc2['Total Day'] == 500  ]
p2_1000 =   sc2.loc[ sc2['Total Day'] == 1000 ]
p2_2000 =   sc2.loc[ sc2['Total Day'] == 2000 ]

p3_50   =   sc3.loc[ sc3['Total Day'] == 50   ]
p3_100  =   sc3.loc[ sc3['Total Day'] == 100  ]
p3_500  =   sc3.loc[ sc3['Total Day'] == 500  ]
p3_1000 =   sc3.loc[ sc3['Total Day'] == 1000 ]
p3_2000 =   sc3.loc[ sc3['Total Day'] == 2000 ]

p4_50   =   sc4.loc[ sc4['Total Day'] == 50   ]
p4_100  =   sc4.loc[ sc4['Total Day'] == 100  ]
p4_500  =   sc4.loc[ sc4['Total Day'] == 500  ]
p4_1000 =   sc4.loc[ sc4['Total Day'] == 1000 ]
p4_2000 =   sc4.loc[ sc4['Total Day'] == 2000 ]
```

```python
p1_avg = [
  p1_50  ['Player 1 Final Score'].mean(),
  p1_100 ['Player 1 Final Score'].mean(),
  p1_500 ['Player 1 Final Score'].mean(),
  p1_1000['Player 1 Final Score'].mean(),
  p1_2000['Player 1 Final Score'].mean()
]

p2_avg = [
  p2_50  ['Player 2 Final Score'].mean(),
  p2_100 ['Player 2 Final Score'].mean(),
  p2_500 ['Player 2 Final Score'].mean(),
  p2_1000['Player 2 Final Score'].mean(),
  p2_2000['Player 2 Final Score'].mean()
]

p3_avg = [
  p3_50  ['Player 3 Final Score'].mean(),
  p3_100 ['Player 3 Final Score'].mean(),
  p3_500 ['Player 3 Final Score'].mean(),
  p3_1000['Player 3 Final Score'].mean(),
  p3_2000['Player 3 Final Score'].mean()
]

p4_avg = [
  p4_50  ['Player 4 Final Score'].mean(),
  p4_100 ['Player 4 Final Score'].mean(),
  p4_500 ['Player 4 Final Score'].mean(),
  p4_1000['Player 4 Final Score'].mean(),
  p4_2000['Player 4 Final Score'].mean()
]
```

```python
fig = plt.subplots(figsize=(15, 10))

players = [50, 100, 500, 1000, 2000] # x


x_ax = np.arange(len(players))

plt.bar(x_ax - 0.2, p1_avg, width = 0.2, align = 'center', label = "Player 1")
plt.bar(x_ax + 0.0, p2_avg, width = 0.2, align = 'center', label = "Player 2")
plt.bar(x_ax + 0.2, p3_avg, width = 0.2, align = 'center', label = "Player 3")
plt.bar(x_ax + 0.4, p4_avg, width = 0.2, align = 'center', label = "Player 4")

plt.xticks(x_ax, ["50", "100", "500", "1000", "2000"])
plt.yticks(
    np.arange(min(min(p1_avg, p2_avg, p3_avg, p4_avg)),
              max(max(p1_avg, p2_avg, p3_avg, p4_avg)),
            1e5))

plt.title("Average Score for Group 8 Players for N-Number of Days")
plt.xlabel('Days')
plt.ylabel('Score (10^6)')
plt.legend()
plt.show()
```

```python
def get_avg_score(df, p):
    p1 = df.loc[ df['Player 1'] == p ]['Player 1 Final Score'].mean()
    p2 = df.loc[ df['Player 2'] == p ]['Player 2 Final Score'].mean()
    p3 = df.loc[ df['Player 3'] == p ]['Player 3 Final Score'].mean() 
    p4 = df.loc[ df['Player 4'] == p ]['Player 4 Final Score'].mean()
    if math.isnan(p1): p1 = 0
    if math.isnan(p2): p2 = 0
    if math.isnan(p3): p3 = 0
    if math.isnan(p4): p4 = 0
    
    return (1/4) * (p1 + p2 + p3 +p4)
```

```python
ovr_avg = []

for i in range(1, 9):
    ovr_avg.append(get_avg_score(df, i))
    
for i in ovr_avg:
    print(i)
    
min(ovr_avg)
```

```python
fig = plt.subplots(figsize=(15, 10))

players = ['Player 1', 'Player 2', 'Player 3', 'Player 4', 'Player 5',
          'Player 6', 'Player 7', 'Player 8'] # x

x_ax = np.arange(len(players))

plt.bar(x_ax, ovr_avg)

plt.xticks(x_ax, players)
plt.yticks(
    np.arange((min(ovr_avg)),
              (max(ovr_avg)),
            1e5)
)

plt.title("Average Score For All Players")
plt.xlabel('Player')
plt.ylabel('Score (10^6)')
plt.legend()
plt.show()
```

```python
def get_avg_score(df, p, r):
    p1 = df.loc[ (df['Player 1'] == p) & (df['Spawn Rate'] == r) ]
    p2 = df.loc[ (df['Player 2'] == p) & (df['Spawn Rate'] == r) ]
    p3 = df.loc[ (df['Player 3'] == p) & (df['Spawn Rate'] == r) ]
    p4 = df.loc[ (df['Player 4'] == p) & (df['Spawn Rate'] == r) ]
    
    l1, l2, l3, l4 = p1.shape, p2.shape, p3.shape, p4.shape
    col1, col2, col3, col4 = l1[1], l2[1], l3[1], l4[1]
    
    if l1[0] == 0: col1 = 0
    if l2[0] == 0: col2 = 0
    if l3[0] == 0: col3 = 0
    if l4[0] == 0: col4 = 0
    
    total_len = (col1 + col2 + col3 + col4)
    s1, s2 = p1['Player 1 Final Score'].sum(), p2['Player 2 Final Score'].sum()
    s3, s4 = p3['Player 3 Final Score'].sum(), p4['Player 4 Final Score'].sum()  
    
    if math.isnan(s1): s1 = 0
    if math.isnan(s2): s2 = 0
    if math.isnan(s3): s3 = 0
    if math.isnan(s4): s4 = 0
    
    return (s1 + s2 + s3 + s4) / (total_len)
```

```python
ovr_avg = []

# array([ 1,  2,  5, 10, 20])

for i in range(1, 9):
    ovr_avg.append(get_avg_score(df, i, 20))
    
fig = plt.subplots(figsize=(15, 10))

players = ['Player 1', 'Player 2', 'Player 3', 'Player 4', 'Player 5',
          'Player 6', 'Player 7', 'Player 8'] # x

x_ax = np.arange(len(players))

plt.bar(x_ax, ovr_avg)

plt.xticks(x_ax, players)
plt.yticks(
    np.arange((min(ovr_avg)),
              (max(ovr_avg)),
            1e5)
)

plt.title("Average Score For All Players at Spawn Rate 20")
plt.xlabel('Player')
plt.ylabel('Score (10^6)')
plt.legend()
plt.show()
```
